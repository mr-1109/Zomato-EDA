```python
# load the country.xlsx file into other dataframe
df_country=pd.read_excel("/Users/rohitpatel/Desktop/Data Science/Zomatodataset/Country-Code.xlsx")
df_country.head()
```

```python
# merge these two dataframe
df_final=pd.merge(df,df_country,on='Country Code',how='left')
df_final.head()
```

```python
df_final.dtypes
```

```python
df_final.columns
```

```python
country_name=df_final.Country.value_counts().index
```

```python
country_val=df_final.Country.value_counts().values
```

```python
# create a pie chart
plt.pie(country_val,labels=country_name)
```

```python
# pie chart for top 3 country

plt.pie(country_val[:3],labels=country_name[:3])
```

```python
# add some % in the pie chart
plt.pie(country_val[:3],labels=country_name[:3],autopct="%1.2f%%")
```

```python
# save this in a variable

ratings=df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating_count'})
```

```python
ratings
```

# observation 
1. when rating is between 4.5 to 4.9 :- food is Excellent 
2. when rating is between 4.0 to 4.4:- food is very good
3. when rating is between 3.5 to 3.9:- food is good
4. when rating is between 2.5 to 3.4:- food is Average
5. when rating is between 1.8 to 2.4:- food is poor


```python
ratings.head()
```

```python
sns.barplot(x='Aggregate rating',y='Rating_count',data=ratings)
```

```python
# increase the figure size
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating_count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating_count',hue="Rating color",data=ratings)
```

```python
# here colors are not matching
## we can use palette

sns.barplot(x='Aggregate rating',y='Rating_count',hue="Rating color",palette=['blue','red','orange','yellow','green','green'],data=ratings)
```

# Observation 
1. not rated count is very high
2. Maximum ratings are between 2.5 to 3.4
